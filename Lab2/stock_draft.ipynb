EXTRACT FROM YFINANCE USING WEB SCRAPING 

```python
symbol = input("Enter stock symbol (e.g. AAPL): ").upper()
date_start = input("Enter start date (m/d/yy): ")
date_end = datetime.today().strftime("%m/%d/%y")
# Convert to Unix timestamps
period1 = int(time.mktime(time.strptime(date_start, "%m/%d/%y")))
period2 = int(time.mktime(time.strptime(date_end, "%m/%d/%y")))


# Set up ChromeDriver path (if not in PATH)
service = Service("./chromedriver")
driver = webdriver.Chrome(service=service)

# Open a page (Yahoo Finance)
url = f"https://finance.yahoo.com/quote/{symbol}/history?period1={period1}&period2={period2}&interval=1d&filter=history&frequency=1d"

driver.get(f"https://finance.yahoo.com/quote/{symbol}/history")

# Let page load
time.sleep(5)

# Get the page source and close the driver
html = driver.page_source
driver.quit()

# Parse with BeautifulSoup
soup = BeautifulSoup(html, "html.parser")

# Example: extract table rows
rows = soup.find_all("tr")
for row in rows[:10]:
    print(row.text)
```

STRUCTURING THE TABLE 

```python
# Create a list to hold row data
data = []

# Parse valid rows
for row in rows:
    cols = row.find_all("td")
    if len(cols) == 7:
        date = cols[0].text.strip()
        open_ = cols[1].text.strip()
        high = cols[2].text.strip()
        low = cols[3].text.strip()
        close = cols[4].text.strip()
        adj_close = cols[5].text.strip()
        volume = cols[6].text.strip()

        data.append([date, open_, high, low, close, adj_close, volume])
        if len(data) == 10:  # limit to first 10 rows
            break

# Define column names
columns = ["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]

# Create DataFrame
df = pd.DataFrame(data,columns=columns)

# Display DataFrame
print(df)
```
